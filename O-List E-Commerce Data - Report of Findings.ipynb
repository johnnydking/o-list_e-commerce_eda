# O-List E-Commerce Data - Report of Findings
The subject of exploration here is real anonymized data from O-List, the largest e-commerce platform in Brazil. The large and rich dataset allows exploration of virtually all aspects of the business. This exploration focuses on sales, product, customer and marketing data.

### Sales by month
We observe total net sales by month across the entirety of the dataset timeline.



Without knowing the parameters of the data sampling used to create the dataset, we cannot be certain of the accuracy of monthly totals, but we see a trend of rapid growth nonetheless. Even with sampled data, O-List reached 974,000 BRL (appx 186,000 USD) in net sales in a single month (11/2017) within roughly one year.

### Sales by category
An overview of product category performance. The sum of net sales by category is observed across the entirety of the dataset timeline, and the categories are sorted in descending order and limited to 10.



The Health and Beauty category is the top performing category with appx 1.2 million BRL in net sales, followed by Watches and Gifts and the Bed, Bath and Table category.

To observe proportionality, these 10 top categories are plotted to represent the percentage of total net sales.



The Health and Beauty accounts for roughly 9.5% of total net sales, followed closely by the Watches and Gifts category at appx 9%.

### Growth by category
Comparing Q1/Q2 2018 to Q1/Q2 2017, the top 10 categories by sales growth are observed, including each category's growth in count of unique skus sold.



The growth of unique skus sold explains the enormous growth of sales in these categories. Some of these categories started with single or double digit sku counts and grew to double or triple digit.

### Average order total and distribution of order totals





The mean order total is 159.94 BRL, and the distribution of order totals is gently right skewed.

### Repeat customers

After calculations, a very low rate of return customers is observed; only 3% of unique customers in the dataset placed more than one order. These return customers accounted for 6% of total orders and 5% of sales. To try and gain insight as to the cause of this low rate of customer return, service related features are viewed for repeat and non-repeat customers to see what the comparison might reveal.

### Growth of seller base
Moving on to the seller base and O-List's marketing campaigns aimed on onboarding new sellers, a count of unique sellers per day is observed.



The seller base displays consistent growth with a clear peak around Christmas of 2017. Next, the impact of marketing campaigns is observed.

### Impact of marketing campaigns



As the total number of sellers onboaded increases, mql sales account for a larger and larger proportion of overall gross sales. The peak is appx 25%. It appears that the marketing campaigns are certainly having a significant amount of success in their contribition to overall revenue.

That said, it's surprising to see such a low number of unique sellers per day relative to how many sellers have been onboarded via marketing campaigns. For example, as the total number of deals closed exceeded 600, there was a peak of only appx 25 unique sellers selling product per day, or 4.2% of onboarded sellers. However, as the total number of onboarded sellers exceeded 750, the number peaks at appx 45 unique sellers, or 6%. This suggests there is a significant lag time from deal close date to first order date.

Still, the percentage of onboarded sellers who are actually selling seems quite low. Starting from the beginning of these marketing campaigns through to September of 2018 (the end of the dataset timeline), only 44% of onboarded sellers had recorded a sale.

### Distribution of sellers by time-to-sales
On average, it takes a seller 52 days to begin selling once onboarded. The distribution of these timeframes is observed.



Of those sellers who do begin selling, most of them record their first sale within 90 days.

### Average revenue per deal over time
The growth of revenue generated over time is observed.



The value of each deal increases substantially over a six month period.

### Average revenue per deal by lead type
The average total revenue generated per deal closed for each lead type, observing the entirety of the dataset timeline.



Unsurprisingly, the 'online big' lead type (well established ecommerce companies) generate by far the most revenue for each deal, but the 'online small' seller should not be discounted.

### Summary and conclusions
The dataset explored here was sampled from the very early days of the O-List platform. The data reveales that roughly a quarter of the company's revenue is generated by its top three categories. Substantial growth of newer categories is observed as new items were added to the site's assortment, and consistent growth of the O-List seller base occurred both organically and as a result of marketing campaigns.

In regards to marketing campaigns, it seems that more follow up and guidance by sales staff on newly closed deals might help increase the number of onboarded sellers who begin to sell, which of course would increase average revenue per deal closed. Observing the time-to-sales distribution combined with revenue over time and revenue per lead type, insight is gained in what type of companies to target, and a reasonable forecast of marketing results becomes possible.
